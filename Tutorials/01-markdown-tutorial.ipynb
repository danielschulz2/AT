## Basic Formatting

### Paragraph Breaks and Line Breaks

```
This is the first paragraph.

This is the second paragraph.
```

- **Line break**: End a line with two spaces

### Tables

Tables use `|` to separate columns and the line below the header defines the alignment:  
`| :-- |` $\to$ left-aligned, `| --: |` $\to$ right-aligned, `| :--: |` $\to$ centered.

```
| We           | love     | NLP           |
|:-------------|:--------:|--------------:|
| left-aligned | centered | right-aligned |
| Natural      | Language | Processing    |
```

| We           | love     | NLP           |
|:-------------|:--------:|--------------:|
| left-aligned | centered | right-aligned |
| Natural      | Language | Processing    |

## Math in Markdown (LaTeX Syntax with `$...$` for inline, `$$...$$` for block)

### Basic Operations

```
Addition: $a + b$  
Subtraction: $a - b$  
Multiplication: $a \cdot b$ or $a * b$  
Division: $\frac{a}{b}$ or $a / b$  
Squared: $x^2$  
Root: $\sqrt{x}$ or $\sqrt[n]{x}$  
Modulo: $a \mod b$  
Equals: $a = b$  
Approximately Equal: $a \approx b$
```

## Images

It may be helpful to insert images in your markdown, especially if you generate figures and want to use them later on in a pen-and-paper task.

First, add the image to the same or a subfolder of your notebook, either by uploading or saving your generated image:

```python
from pandas import DataFrame
import seaborn as sns
import matplotlib.pyplot as plt

example_df = DataFrame({"sentence_id": [1,2,3,4,5], "length": [5,4,7,9,6]})
sns.barplot(example_df, x="sentence_id", y="length").set(xlabel="Sentence", ylabel="Word Count")
plt.savefig("example.svg", format="svg")
```

Afterwards, you have two ways of inserting the image into your markdown cell. 
1. You can use the markdown syntax for links starting with an exclamation mark in a new line, which renders the image at its original size:
   
   `![example image](example.svg)`

   ![example image](example.svg)
2. If you need more control over the image size, e.g., because you want to insert two images side by side, you can incorporate HTML syntax inside your markdown cell:
   
   ```html
   <div>
       <img src="example.svg" alt="example image" width="400"/>
       <img src="example.svg" alt="example image" width="400"/>
   </div>
   ```
    <div>
        <img src="example.svg" alt="example image" width="400"/>
        <img src="example.svg" alt="example image" width="400"/>
    </div>
3. *Bonus tip* If you do not want to upload as many images to the JupyterHub and your images are relatively small, you can also convert them to a Base64 text representation [online](https://www.base64-image.de/tutorial) and add this text snippet to `src` tag in method 2. However, you should only do this with very small images, since the final string is very long and makes the Jupyter Notebook and text editors very laggy for larger images:
   
   `<img src="data:image/svg+xml;base64,PD94bW[...]ZnPg0K" width="400" />`
